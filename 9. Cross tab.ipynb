# 13. Cross-Tab

```python
import pandas as pd
df=pd.read_excel("survey.xlsx")
df

# crosstab - pandas
```

```python
pd.crosstab(df.Nationality,df.Handedness,margins=True) ##showing the frequency of the variable
#margins shows the total count
```

```python
pd.crosstab(df.Sex,[df.Handedness,df.Nationality],margins=True) #col,row
```

```python
pd.crosstab([df.Nationality,df.Sex],[df.Handedness],margins=True) 
```

```python
pd.crosstab([df.Sex],[df.Handedness],margins=True) # hover over the crosstab function, click shift+tab, it show the documentation
```

```python
pd.crosstab([df.Sex],[df.Handedness],margins=True,normalize="index") #normalize shows the percentage
```

```python
import numpy as np
pd.crosstab([df.Sex],[df.Handedness],values=df.Age,aggfunc=np.average)
```
