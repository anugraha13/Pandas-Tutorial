#  Handle missing data - Replace()

```python
import pandas as pd
df=pd.read_csv("missing_data_replace.csv")
df
```

```python
import numpy as np
new_df=df.replace([-99999,-8888],np.NaN)
new_df
```

```python
df
```

```python
new_df=df.replace({
    'temperature':-99999,
    'windspeed':-99999,
    'event':"No Event"
},np.NaN)
new_df
```

```python
new_df=df.replace({
    -99999:np.NaN,
    'No Event':"Sunny"
})
new_df
```

```python
df
```

```python
new_df=df.replace('[A-Za-z]','',regex=True)
new_df
```

```python
new_df=df.replace(
{
    'temperature':'[A-Za-z]',
    'windspeed':'[A-Za-z]'
},'',regex=True)
new_df
```
