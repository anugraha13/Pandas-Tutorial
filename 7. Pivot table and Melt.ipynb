# 10. Pivot Table

1. index, columns, values

```python
import pandas as pd
df=pd.read_csv("weather_by_cities.csv")
df
```

```python
df.pivot(index="day",columns="city",values="temperature")
```

```python
df.pivot(index="temperature",columns="city")
```

```python
df.pivot_table(index="city",columns="day")#used to summarize and aggreate data inside the data frame
```

```python
df.pivot_table(index="city",columns="day",aggfunc="count")
```

```python
df.pivot_table(index="city",columns="day",margins=True)
```

```python
import pandas as pd
df=pd.read_csv("Groupper.csv")
df
```

```python
# to_datetime
# grouper - pandas
# grouper(freq,key)

df['date']=pd.to_datetime(df['date'])
```

```python
df.pivot_table(index=pd.Grouper(freq='M',key="date"),columns="city")
```

# 11. Reshape dataframe using Melt()

```python
# melt - pandas function

import pandas as pd
df=pd.read_csv("Melt.csv")
df
```

```python
#id_vars 

df1=pd.melt(df,id_vars=["day"])
df1
```

```python
df1[df1["variable"]=="temperature"]
df1
```

```python
df1=pd.melt(df,id_vars=["day"],var_name="category",value_name="Humidity")
df1
```
